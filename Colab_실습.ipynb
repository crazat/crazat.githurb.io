## groupby 사용하기
- 데이터를 그룹화하여 집계해서 가공하는데 아주 중요하다.
- 전체 데이터를 지정한 컬럼 기준으로 그룹 별로 나누고, 각 그룹별로 집계함수를 적용한 후, 그룹별 집계 결과를 하나로 합치는 단계를 거치게 된다.

20221222142824.png

- groupy by 결과에 대한 단일 컬럼을 지정하여 집계하는 예시
    - 항구별 생존률을 집계하고 싶다면??

````python
df.groupby("탑승항구")["생존"].mean()  # 시리즈 반환
````

````python
df.groupby("탑승항구")["생존"].agg("mean") # agg 메소드 사용하여 집계 종류를 문자열로 전달 하는 예시
````

- groupy by 결과에 대해 agg 메소드에 딕셔너리를 전달하여 집계하는 예시.
    

````python
agg_dict = {"생존":"mean"} # key는 컬럼, value는 집계함수 혹은 집계 문자열
df.groupby("탑승항구").agg(agg_dict) # 데이터 프레임 형태로 반환된다.
````

- 여러 컬럼 집계하기

## 데이터 프레임 병합하기

- merge
    - 두개의 데이터 프레임을 기준이 되는 컬럼명의 같은 값을 기준으로 합친다.
    - how
        - inner(기본값): 기준이 되는 컬럼의 데이터로 merge 하는데 같은 값이 아닌 행들은 사라져서 반환된다.
        - left: 기준이 되는 컬럼의 데이터로 merge 하는데 왼쪽 데이터 프레임의 같은 값이 아닌 행들이 사라지지 않고 반환된다.

    ```python
    pd.merge(left, right, on='기준 컬럼', how='left')
    ```

- 머신러닝 모델에 학습데이터는 모두 수치형을 넣어줘야한다.
- 객실번호는 문자열이다. 객실번호를 의미 있는 수치형으로 변경하기위해 다음과 같이 전처리를 해보자
    - 객실번호별 운임료의 평균을 집계하여 데이터프레임 생성
    - merge 함수를 이용하여 각 객실번호에 해당하는 운임료평균 컬럼을 추가

````python
# 객실번호별 운임료 평균 집계
agg_dict = {"요금":"mean"}
tmp = df.groupby("객실번호").agg(agg_dict)
tmp
````

````python
tmp = tmp.reset_index() # 인덱스에 있는 객실번호를 컬럼 으로 만들기
tmp
````

````python
# 데이터에 동일한 요금 컬럼이 있어 컬럼명 변경
tmp = tmp.rename(columns= {"요금" : "객실_운임료_평균"})
tmp
````

````python
df = pd.merge(df, tmp, how='left', on="객실번호" ) # on 생략시 동일한 이름의 컬럼을 기준으로 머지한다.
df
````
