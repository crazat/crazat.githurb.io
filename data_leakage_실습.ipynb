# Data leakage
- 머신러닝에서 Data leakage는 예측 시 사용할 수 없는 정보가 모델의 학습및 평가에서 사용되는 것을 말한다.
- Data leakage 문제가 있으면 모델이 과대평가 되고 모델의 일반화 가능성이 보장되지 않는다.
- Data leakage 문제
    - 실제 서비스 환경에서 예측할 때 사용할 수 없는 피처를 수집하여 학습 데이터와 테스트 데이터에서 사용하여 모델 학습 및 성능을 평가 하는 경우
    - 테스트 데이터의 정보를 모델의 학습및 평가에 사용하는 경우


# 테스트 데이터의 정보를 모델의 학습및 평가에 사용하는 경우
- 우리가 모르는 데이터라고 가정한 테스트 데이터의 정보를 모델의 학습에 사용할 경우 Data leakage 이다.

## 다른 샘플들의 통계치를 이용하여 피처를 생성하는 경우
- "범주4" 컬럼 별로 "수치1" 컬럼의 평균값을 피처로 추가하기

```python
agg_dict = {"수치1":"mean"}
train_agg = train_customer.groupby("범주4").agg(agg_dict).add_suffix("_mean").reset_index()
train_agg
```

- merge

```python
train_ft = pd.merge(train_ft,train_agg,how="left",on="범주4")
train_ft.head()
```

### Data leakage 발생 예시
- 테스트 데이터를 집계하여 피처를 추가할 경우 Data leakage 이다.

```python
test_agg = test_customer.groupby("범주4").agg(agg_dict).add_suffix("_mean").reset_index()
pd.merge(test_ft,test_agg,how="left",on="범주4")
```

### Data leakage 해결 방안
- 학습데이터에서 집계한 train_agg 변수로 merge 해야한다.

```python
test_ft = pd.merge(test_ft,train_agg,how="left",on="범주4")
test_ft
```

### Data leakage가 아닌 경우
- 변제내역 데이터에서 고객별 연체횟수 합계 피처 생성 후 merge

```python
# 학습데이터
agg_dict = {"연체횟수유형1":"sum"}
tmp = train_payment.groupby("고객ID").agg(agg_dict).add_suffix("_sum").reset_index()
pd.merge(train_ft,tmp,how="left",on="고객ID")
```

- 하나의 샘플이 여러 개의 행으로 구성된 데이터 집계하여 피쳐생성할 때 테스트데이터에도 동일한 방식으로 집계하는 것은 Data leakage가 아니다.

```python
# 테스트 데이터
tmp = test_payment.groupby("고객ID").agg(agg_dict).add_suffix("_sum").reset_index()
pd.merge(test_ft,tmp,how="left",on="고객ID")
```

- 다른 샘플을 활용한 것이 아니기 때문

```python
test_payment.groupby("고객ID").apply(lambda x : x.set_index("고객ID")).head(50)
```

- 변제 내역 데이터에서 피처 추출해서 merge

```python
agg_dict = {
    # key: 집계대상컬럼
    # value: 변경할 컬럼명과 집계문자열/함수를 튜플에 담고 리스트에 넣으면 된다.
    "연체횟수유형1": [
        ("연체횟수유형1_sum","sum"),
        ("변제횟수","count") # 카운팅하여 변제횟수 피쳐 생성
    ],
    "연체횟수유형2": [
        ("연체횟수유형2_sum","sum")
    ],
    "연체횟수유형3": [
        ("연체횟수유형3_sum","sum")
    ],
    "총연체일수": [
        ("총연체일수_sum","sum")
    ],
    "정상결제횟수": [
        ("정상결제횟수_sum","sum"),
        ("정상결제횟수_mean","mean"),
        ("정상결제횟수_std","std"),
    ],
    "카드코드": [
        ("카드코드_nunique","nunique"),
    ],
    "카드한도": [
        ("카드한도_mean","mean"),
        ("카드한도_std","std"),
    ],
    "카드현재잔액": [
        ("카드현재잔액_mean","mean"),
        ("카드현재잔액_std","std"),
    ],
    "카드최고잔액": [
        ("카드최고잔액_mean","mean"),
        ("카드최고잔액_std","std"),
    ],
}
tmp = train_payment.groupby("고객ID").agg(agg_dict)
tmp # 멀티 컬럼이 생성됨
```

```python
# 멀티컬럼 제거하고 고객ID 컬럼데이터로 만들기
tmp.columns = tmp.columns.droplevel() # 기본으로 0레벨 컬럼 제거해서 반환
tmp = tmp.reset_index()  # 고객ID 컬럼데이터로 만들기
tmp
```

```python
# 추출한 피쳐 merge
train_ft = pd.merge(train_ft,tmp,how="left",on="고객ID")
train_ft.head()
```

```python
# 테스트 데이터도 동일한 방식으로 피쳐 추가
tmp = test_payment.groupby("고객ID").agg(agg_dict) # 동일한 방식으로 집계
tmp.columns = tmp.columns.droplevel() # 기본으로 0레벨 컬럼 제거해서 반환
tmp = tmp.reset_index()  # 고객ID 컬럼데이터로 만들기
test_ft = pd.merge(test_ft,tmp,how="left",on="고객ID")
test_ft.head()
```

### Data leakage 발생 예시

- 원핫 인코딩을 할 때 테스트 데이터에 대하여 fit 을 하는 경우 Data leakage 이다.

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
tmp = pd.DataFrame(
    enc.fit_transform(test_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)
pd.concat([test_ft, tmp], axis=1)
```

### Data leakage 해결 방안
- 테스트 데이터에는 `transform` 만 해야한다. `fit` 하면 안된다.
- 모르는 데이터라고 가정한 테스트 데이터에 학습 데이터에는 없는 새로운 범주가 나왔을 경우 무시하기 위해 원핫 인코더에 `handle_unknown='ignore'` 인수를 주는 것이다.

```python
enc = OneHotEncoder(handle_unknown='ignore')
# 학습데이터
tmp = pd.DataFrame(
    enc.fit_transform(train_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)
train_ft = pd.concat([train_ft, tmp], axis=1)

# 테스트 데이터
tmp = pd.DataFrame(
    enc.transform(test_ft[cols]).toarray(), # transform만 해야한다.
    columns = enc.get_feature_names_out()
)
test_ft = pd.concat([test_ft, tmp], axis=1)

train_ft.shape , test_ft.shape
```

# Data leakage로 인한 모델의 과대 평가 예시

```python
SEED = 42
```

- 데이터 불러오기

```python
train = pd.read_csv(f"{DATA_PATH}titanic_train.csv") # 학습데이터
test = pd.read_csv(f"{DATA_PATH}titanic_test.csv") # 테스트 데이터
test_target = pd.read_csv(f"{DATA_PATH}titanic_test_target.csv") # 테스트데이터 정답값
train.shape , test.shape , test_target.shape
```

- 결측치 확인하기

```python
train.isnull().sum()
```

- 학습 데이터에서 얻은 통계량을 이용하여 결측치를 채워야한다.

```python
age_mean = train["age"].mean() # 평균
fare_median = train["fare"].median() # 중앙값
cabin_unk = "UNK" # 새로운 범주
embarked_mode = train["embarked"].mode()[0] # 최빈값
age_mean , fare_median ,cabin_unk , embarked_mode
```

- 학습데이터 결측치 처리

```python
train["age"] = train["age"].fillna(age_mean)
train["cabin"] = train["cabin"].fillna(cabin_unk)
```

- 테스트데이터 결측치 처리

```python
test["age"] = test["age"].fillna(age_mean)
test["fare"] = test["fare"].fillna(fare_median)
test["cabin"] = test["cabin"].fillna(cabin_unk)
test["embarked"] = test["embarked"].fillna(embarked_mode)
```

```python
train.isnull().sum().sum() , test.isnull().sum().sum()
```

- 특성으로 사용할 변수 추가하기

```python
cols = ["age","sibsp","parch","fare","pclass","gender","embarked","cabin"]
train_ft = train[cols].copy()
test_ft = test[cols].copy() # 테스트 데이터
train_ft.shape, test_ft.shape
```

- 범주형 변수 원핫인코딩하여 특성으로 추가하기

```python
cols = ['gender','embarked']
enc = OneHotEncoder(handle_unknown = 'ignore') # 모르는 범주가 있을 경우 무시
enc.fit(train_ft[cols])
```

```python
# 학습 데이터
tmp = pd.DataFrame(
    enc.transform(train_ft[cols]).toarray(), # ndarray
    columns = enc.get_feature_names_out() # 컬럼명
)
train_ft = pd.concat([train_ft,tmp],axis=1)
train_ft.head()
```

```python
# 테스트 데이터
tmp = pd.DataFrame(
    enc.transform(test_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)
test_ft = pd.concat([test_ft,tmp],axis=1)
test_ft.head()
```

## cabin 별 fare와 age의 평균값을 피처로 추가

```python
agg_dict = {"fare": "mean","age": "mean"}
tmp = train.groupby("cabin").agg(agg_dict).add_suffix("_cabin_mean").reset_index()
train_ft = pd.merge(train_ft,tmp,how="left",on="cabin")
train_ft.head()
```

### Data leakage에 해당하는 테스트 셋

- 위에서 추가한 피처 제외

```python
test_ft.iloc[:,:-2].head()
```

- 테스트 데이터를 집계하여 피처를 추가해서 Data leakage 문제 발생

```python
tmp = test.groupby("cabin").agg(agg_dict).add_suffix("_cabin_mean").reset_index()
test_ft_leakage = pd.merge(test_ft.iloc[:,:-2],tmp,how="left",on="cabin")
test_ft_leakage.head()
```

- 파생변수 생성과정에서 생긴 결측치 처리하기

```python
train_ft.isnull().sum().sum(), test_ft.isnull().sum().sum(), test_ft_leakage.isnull().sum().sum()
```

```python
test_ft.isnull().sum()
```

```python
# 학습데이터에서 얻은 통계치로 결측치를 처리
cols = ["fare_cabin_mean","age_cabin_mean"]
median_series = train_ft[cols].median()
test_ft[cols] = test_ft[cols].fillna(median_series)
test_ft.isnull().sum().sum()
```

- 문자열 데이터 제거

```python
cols = ["gender","embarked","cabin"]
train_ft = train_ft.drop(columns=cols)
test_ft = test_ft.drop(columns=cols)
test_ft_leakage = test_ft_leakage.drop(columns=cols) # data leakage 데이터
```

- 정답 데이터

```python
target = train["survived"]
target
```

- 모델 교차검증 점수 확인하기

```python
scores = cross_val_score(model,train_ft,target,cv = cv ,scoring='roc_auc',n_jobs = -1)
np.mean(scores)
```

- 모델 학습

```python
model.fit(train_ft,target)
```

## Data leakage에 아닌 테스트 데이터 예측 및 평가

```python
from sklearn.metrics import roc_auc_score
y_test = test_target["survived"] # 테스트셋 y값
pred = model.predict_proba(test_ft)[:,1] # 예측
roc_auc_score(y_test,pred) # AUC 평가
```

- 모델이 과대평가 된다.

```python
y_test = test_target["survived"] # 테스트셋 y값
pred = model.predict_proba(test_ft_leakage)[:,1] # 예측
roc_auc_score(y_test,pred) # AUC 평가
```
